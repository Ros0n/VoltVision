```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
timeMap = {
    f'H{i+1}': f'{i:02d}' for i in range(0, 24)
}
```

```python
df_price["Time"]=df_price["Hours"].apply(lambda x: timeMap[x])
```

```python
df_price[df_price['Hours'].isna()==True]
```

```python
df_priceNew = df_price.dropna(how='all')
df_priceNew
```

```python
df_priceNew[df_priceNew['Hours'].isna()==True]
```

```python
df_priceNew.loc[:, "Time"] = df_priceNew["Hours"].apply(lambda x: timeMap[x] + ":00:00")
```

```python
df_priceNew
```

```python
df_priceNew.loc[:, 'Datetime'] = df_priceNew['Delivery Day'] + ' ' + df_priceNew['Time']
```

```python
df_priceNew
```

```python
df_priceNew.loc[:, 'Datetime'] = pd.to_datetime(df_priceNew['Datetime'])
```
